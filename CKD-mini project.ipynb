# logistic regression

```python
from sklearn.model_selection import train_test_split 
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score, precision_score,recall_score,f1_score
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
data3.info()
```

```python
x=data3.drop(['ckd'],axis=1)
```

```python
y=data3['ckd']
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.25,random_state=25)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
logit_model = LogisticRegression(class_weight={0:0.4, 1:1})

logit_model.fit(x_train,y_train)

y_pred=logit_model.predict(x_test)

y_pred.shape
```

```python
print("accuracy is", accuracy_score(y_test,y_pred))
print("Precision is", precision_score(y_test,y_pred))
print("Recall is", recall_score(y_test,y_pred))
print("f1 is", f1_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
y_pred_proba=logit_model.predict_proba(x_test)[:,1]

y_pred_2=(y_pred_proba> 0.65).astype(int)

accuracy_score(y_test, y_pred_2)
```

```python
print("Precision is", precision_score(y_test,y_pred_2))
print("Recall is", recall_score(y_test,y_pred_2))
print("f1 is", f1_score(y_test,y_pred_2))
```
